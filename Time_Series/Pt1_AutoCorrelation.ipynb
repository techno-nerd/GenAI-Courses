```python
import pandas as pd
import numpy as np
```

```python
sales = pd.DataFrame({'Month':range(180),'Base':0,'Promotion':0})
sales.set_index('Month', inplace=True)
sales.head()
```

```python
#Base sales = 1000 / month

sales['Base'] = 1000 + np.random.randint(-10, 10, 180)
sales['Base'].plot()
```

```python
#Promotion once every 5 months (starting 6th month), only impacts that month

sales.loc[5, 'Promotion'] = 40 #40 additional sales

for i in range(10, 180, 5):
    sales.loc[i, 'Promotion'] = sales.loc[i-5, 'Promotion'] + np.random.randint(-2,2)

sales['Promotion'].plot()
```

```python
sales['Total'] = sales.Base + sales.Promotion
sales.Total.plot()
```

```python
#Auto-correlation and partial auto-correlation plots

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(sales.Total)
plot_pacf(sales.Total);
```
